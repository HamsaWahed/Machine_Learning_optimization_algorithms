#### Use your final version of the implemented Batch GD algorithm for single variable to find the best line filt of the given data.
#### Choose the suitable number of iterations, learning rate, and stop criteria.
#### Calculate r2 score. Shouldn't below 0.9
#### Plot the required curves (loss-epochs, loss-theta0, loss-theta1, all fitted lines per epoch (single graph), best fit line)

```python
def GD(X,y,alpha,iter_max):
    
    theta_0,theta_1 =0,0
   
    
    for i in range(iter_max):
        print(f'****************** iteration{i}**************************')
        yhat = theta_0 + theta_1 * X
        lines.append(yhat)
        print("h(x):",yhat)
        m = len(X)
        e = yhat - y
        print("Error Vector:\n",e)
        J = (e @ e)/(2*m)
        print("j = ",J)
        loss.append(J)
        d_theta_0 = np.sum(e)/m
        d_theta_1 = (e @ X)/m
        print("d_theta_0 : ",d_theta_0)
        print("d_theta_1 : ",d_theta_1)
        theta_0 = theta_0 - (alpha * d_theta_0)
        th0.append(theta_0)

        theta_1 = theta_1 - (alpha * d_theta_1)
        th1.append(theta_1)
        
        print("theta_0_new : ",theta_0)
        print("theta_1_new : ",theta_1)
        Grad_norm = np.sqrt(((d_theta_0)**2 + (d_theta_1)**2))
        print("Gradient Norm = ",Grad_norm)
        print("predicted Y: \n",yhat)
        if i != 0 and (Grad_norm < 0.001 or abs(loss[-1] - loss[-2]) < 0.001):
            break
        
```

### Update your Batch GD for single variable  implementation (Batch variant) to be Momentum-Based GD and check your results

####  takes values between 0 and 1.

```python
def MGD(X,y,alpha,iter_max,gamma):
    
    theta_0,theta_1 =0,0
    vo0,vo1 = 0,0
    
    for i in range(iter_max):
        print(f'****************** iteration{i}**************************')
        yhat = theta_0 + theta_1 * X
        lines.append(yhat)
        print("h(x):",yhat)
        m = len(X)
        e = yhat - y
        print("Error Vector:\n",e)
        J = (e @ e)/(2*m)
        print("j = ",J)
        loss.append(J)
        d_theta_0 = np.sum(e)/m
        d_theta_1 = (e @ X)/m
        print("d_theta_0 : ",d_theta_0)
        print("d_theta_1 : ",d_theta_1)
        
        vo0 = (gamma*vo0) + (alpha*d_theta_0)
        vo1 = (gamma*vo1) + (alpha*d_theta_1)
        
        print("vo0 : ",vo0)
        print("vo1 : ",vo1)
        
        
        theta_0 = theta_0 - vo0
        th0.append(theta_0)

        theta_1 = theta_1 - vo1
        th1.append(theta_1)
        
        print("theta_0_new : ",theta_0)
        print("theta_1_new : ",theta_1)
        Grad_norm = np.sqrt(((d_theta_0)**2 + (d_theta_1)**2))
        print("Gradient Norm = ",Grad_norm)
        print("predicted Y: \n",yhat)
        if i != 0 and (Grad_norm < 0.001 or abs(loss[-1] - loss[-2]) < 0.001):
            break
        
```

### Update your Batch GD for one variable  implementation to be NAG and check your results

##### Very Important:
- <b><font color='red'> In order to calculate the gradient at theta_temp, you will need to calculate the prediction (hypothesis output) using theta_temp. </font></b> 
##### h_temp = theta0_temp + theta1_temp * X 

```python
def NAG_GD(X,y,alpha,iter_max,gamma):
    
    theta_0,theta_1,temp0,temp1=0,0,0,0
    vo0,vo1 = 0,0
    
    for i in range(iter_max):
        print(f'****************** iteration{i}**************************')
        yhat = temp0 + temp1 * X
        lines.append(yhat)
        
        print("h(x):",(theta_0 + theta_1* X))
        
        m = len(X)
        e = yhat - y
        e_loss = (theta_0 + theta_1* X) - y
        print("Error Vector:\n",e_loss)
        
        J = (e_loss @ e_loss)/(2*m)
        print("j = ",J)
        loss.append(J)
        
        d_temp0 = np.sum(e)/m
        d_temp1 = (e @ X)/m
        
        
        vo0 = (gamma*vo0) + (alpha*d_temp0)
        vo1 = (gamma*vo1) + (alpha*d_temp1)
        
        temp0 = theta_0 - gamma*vo0
        temp1 = theta_1 - gamma*vo1
        
        print("vo0 : ",vo0)
        print("vo1 : ",vo1)
        
       
        th0.append(theta_0)
        th1.append(theta_1)
        
        print("theta_0_new : ",theta_0)
        print("theta_1_new : ",theta_1)
    
        Grad_norm = np.sqrt(((d_temp0)**2 + (d_temp1)**2))
        print("Gradient Norm = ",Grad_norm)
        print("predicted Y: \n",yhat)
        if i != 0 and (Grad_norm < 0.001 or abs(loss[-1] - loss[-2]) < 0.001):
            break
            
        theta_0 = temp0 - (alpha*d_temp0)
        theta_1 = temp1 - (alpha*d_temp1)
```
